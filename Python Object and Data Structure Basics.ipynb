# Numbers

### Operators: { +, -, /, *, %, **, // }
    - + : addition<br>
    - -  : subtraction<br>
    - /  : true division<br>
    - *  : multiplication<br>
    - %  : modulo<br>
    - ** : power<br>
    - // : int division<br>
    

```python
print(2 ** 3)
```

```python
print(1 / 2)
```

```python
print(1 // 2)
```

### Lists

    - Mutable
    - Ordered Sequence
    - Mutation is in place (void return type)
    

```python
myList = ['one', 2, 3.0]
anotherList = ['hello world']
combinedList = myList + anotherList
myList.append('3')
appendedList = myList
print(combinedList)
print(appendedList)
```

```python
letters = ['a','r','t','h','u','u','y','er']
number = [3,4,5,123,23,32,1,1,5,8,1,99]
number1 = ['3','4','5','123','23','32','1','1','5','8','1','99']
letters[7] = '3'
letters.sort()
number.sort()
number1.sort()
print(letters)
print(number) 
print(number1)
```
